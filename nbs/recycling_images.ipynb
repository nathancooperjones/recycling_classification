```python
from io import BytesIO
import os
import time

from fastai.vision import *
from fastai.vision import image as im
from fastai.metrics import accuracy
from PIL import Image
import requests
```

```python
np.random.seed(42)

bs = 64  # can go lower if out of GPU memory
```

```python
cd /home/ubuntu/recycling_classification/
```

## Looking at the data

```python
image_path = Path('/home/ubuntu/images')
```

```python
from recycling_classification.prepare_data import validate_images
```

```python
validate_images(image_path, ext='jpg')
```

```python
reg_ex_pat = r'/([^/]+)_\d+.jpg$'
```

```python
transforms = get_transforms(do_flip=True,
                            flip_vert=True,
                            max_rotate=10.0,
                            max_zoom=1.2,
                            max_lighting=0.5,
                            max_warp=0.35,
                            p_affine=0.8,
                            p_lighting=0.75)
```

```python
data = ImageDataBunch.from_name_re(image_path,
                                   file_names,
                                   reg_ex_pat,
                                   ds_tfms=transforms,
                                   size=224,
                                   bs=bs
                                   ).normalize(imagenet_stats)
```

```python
data.show_batch(rows=3, figsize=(12,10))
```

```python
assert len(data.classes) == data.c
data.classes
```

## Training: resnet50

```python
learn = cnn_learner(data, models.resnet50, metrics=accuracy)
```

```python
learn.fit_one_cycle(4)
```

```python
learn.save('stage-1')
```

## Results

```python
interp = ClassificationInterpretation.from_learner(learn)

losses, idxs = interp.top_losses()

assert len(data.valid_ds)==len(losses)==len(idxs)

interp.plot_top_losses(9, figsize=(15,11))
```

```python
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```

```python
interp.most_confused(min_val=2)
```

## Unfreezing, fine-tuning, and learning rates

```python
learn.load('stage-1');
```

```python
learn.lr_find()
```

```python
learn.recorder.plot(suggestion=True)
```

```python
learn.unfreeze()
learn.fit_one_cycle(2, max_lr=slice(5e-5, 1e-5))
```
